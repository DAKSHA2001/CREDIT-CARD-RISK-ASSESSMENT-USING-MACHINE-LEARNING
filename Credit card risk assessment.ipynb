```python
credit = credit.drop(["ID"], axis=1)
```

```python
credit.head(5)
```

```python
credit.rename(columns={'PAY_0':'PAY_1'}, inplace=True)
```

```python
credit.head(5)
```

```python
credit['EDUCATION'].value_counts()
```

```python
credit["EDUCATION"] = credit["EDUCATION"].map({0:4,1:1,2:2,3:3,4:4,5:4,6:4})
credit["MARRIAGE"] = credit["MARRIAGE"].map({0:3,1:1,2:2,3:3})
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = credit.drop(['default.payment.next.month'], axis=1)
X = scale.fit_transform(X)
```

```python
Y = credit['default.payment.next.month']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=1)
```

```python
model.fit(X_train,Y_train)
```

```python
model.fit(X_test,Y_test)
model.score(X_test,Y_test)
```

```python
model.predict(X_test)
```

```python
x_demo=[[5000,2,2,2,64,0,0,2,2,2,2,1000,500,2000,500,600,700,0,0,1000,500,600,700]]
if model.predict(x_demo)==0:
    print("Person is likely to NOT have default payment next month")
else:
    print("Person is likely to have default payment next month")
```
